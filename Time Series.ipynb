### importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import calendar
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

## Import Local Dataset: Portlands Unemployment Rate 2003-2023

```python
# Import the dataset
portland_unemp_rate = pd.read_csv('PORT941UR.csv')
```

### Cleaning dataset (missing values)

```python
# finding missing values
missing_values = portland_unemp_rate.isnull().sum()  # Count missing values in each column
print(missing_values)
```

### Result: We dont have any missing values in this dataset

### Rename columns

```python
portland_unemp_rate.columns = ['Date', 'Portlands Unemployment Rate']
portland_unemp_rate
```

### Converting the Date column to the datetime format
### Highest Portland's unemployment Rate 

```python
# Convert 'Date' column to datetime
portland_unemp_rate['Date'] = pd.to_datetime(portland_unemp_rate['Date'])

# seperate month and year
portland_unemp_rate['Month'] = portland_unemp_rate['Date'].dt.month
portland_unemp_rate['Year'] = portland_unemp_rate['Date'].dt.year
portland_unemp_rate['Month'] = portland_unemp_rate['Month'].apply(lambda x: calendar.month_abbr[x])  # Change month number to its name

# Finding and printing the highest unemployment rate with its corresponding year and month
highest_rate_row = portland_unemp_rate.loc[portland_unemp_rate['Portlands Unemployment Rate'].idxmax()]
highest_rate = highest_rate_row['Portlands Unemployment Rate']
highest_year = highest_rate_row['Year']
highest_month = highest_rate_row['Month']
print(f"\n\n\nHighest Portland's unemployment Rate was: {highest_rate}% in {highest_month} {highest_year}\n\n\n")
```

# Plot 1: Plot with label=month.

```python
# Plot 1
plt.figure(figsize=(12, 8))  # Adjusted figure size for better visualization
months_order = calendar.month_abbr[1:]  # List of abbreviated month names in order

# Using a for loop
for month in months_order: 
    month_data = portland_unemp_rate[portland_unemp_rate['Month'] == month]
    sns.lineplot(x='Year', y='Portlands Unemployment Rate', data=month_data, label=month)

# Formatting the plot
plt.title('Portlands Unemployment Rate Across Months')
plt.xlabel('Year')
plt.ylabel('Portlands Unemployment Rate (%)')
plt.legend(title='Month')
plt.grid(True)
plt.tight_layout()
# Show plot
plt.show()
```

# Result of plot 1 : 
## <span style="color:orange;">How has Portland's unemployment rate changed over the last twenty years.</span>
### The Portlands unemployment rate illustrates a fluctuating pattern over time. From 2003 to 2006, it steadily declined. However, between 2006 and 2010, there was a noticeable increase. A decreasing trend in unemployment rate emerged from 2010 to 2019, just before the onset of the Covid-19 pandemic. Unfortunately, during of Covid-19 pandamic, the Portlands unemployment rate surged in 2020. But this rete gradually declined as we progressed towards 2023.

```python
# combining the year and month together to a new column
usa_unemp_rate['Date'] = pd.to_datetime(usa_unemp_rate['Year'].astype(str) + '-' + usa_unemp_rate['Period'].str[1:], format='%Y-%m')

# changing the name of columns
usa_unemp_rate.rename(columns={'Value': 'USA Unemployment Rate'}, inplace=True)

# removing the year and period columns
usa_unemp_rate.drop(['Year', 'Period'], axis=1, inplace=True)

usa_unemp_rate.head()
```

# Smoothing Techniques

# 1. Smoothed Plot (Moving Average)

```python
window_size = 20 # size of smoothing, higher means more smoothing

# Smooth portland_unemp_rate
portland_unemp_rate['Smoothed Portland Rate'] = portland_unemp_rate['Portlands Unemployment Rate'].rolling(window=window_size).mean()

# Smooth usa_unemp_rate
usa_unemp_rate['Smoothed USA Rate'] = usa_unemp_rate['USA Unemployment Rate'].rolling(window=window_size).mean()

# Plot the smooth time series
plt.figure(figsize=(10, 6))

# Plot portland_unemp_rate
plt.plot(portland_unemp_rate.index, portland_unemp_rate['Smoothed Portland Rate'], label="Smoothed Portland's Unemployment Rate", color='green')

# plot usa_unemp_rate
plt.plot(usa_unemp_rate.index, usa_unemp_rate['Smoothed USA Rate'], label='Smoothed USA Unemployment Rate', color='blue')

# Format plot
plt.title('Smoothed Comparison of Portland and USA Unemployment Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (Smoothed)')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

## Result of Smoothed Plot (Moving Average):
###  As you can see, the two time series are smoother than before. <span style="color:orange;">However, why did we see a higher unemployment rate during 2009 to 2011 compared to 2020-2021 in this smoothed plot?</span> The reason lies in the differences in trends between these two periodsâ€”the characteristics of unemployment trends during the COVID-19 pandemic in April 2020 and the recession in 2009 are totally different. A rapid, sharp increase followed by a quick decline in April 2020 will be smoothed differently than a slower, more prolonged increase and decrease in 2009-2011. This discrepancy is why, while smoothing aids in visualizing trends, for precise analysis or identifying extreme values, referring back to the original unsmoothed data becomes necessary. <span style="color:orange;">What can we do to address this?</span>

# 2. Exponential Smoothed Plot

```python
alpha_portland = 0.1 
portland_unemp_rate['Exponential_Smoothed_Portland'] = portland_unemp_rate['Portlands Unemployment Rate'].ewm(alpha=alpha_portland, adjust=False).mean()

alpha_usa = 0.1 
usa_unemp_rate['Exponential_Smoothed_USA'] = usa_unemp_rate['USA Unemployment Rate'].ewm(alpha=alpha_usa, adjust=False).mean()

plt.figure(figsize=(10, 6))

plt.plot(portland_unemp_rate.index, portland_unemp_rate['Exponential_Smoothed_Portland'], label="Smoothed Portland's Unemployment Rate", color='green')

plt.plot(usa_unemp_rate.index, usa_unemp_rate['Exponential_Smoothed_USA'], label='Smoothed USA Unemployment Rate', color='blue')

plt.title('Comparison of Exponentially Smoothed Portland and USA Unemployment Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

## Result of Exponential Smoothed Plot:
### The exponential smoothed plot demonstrates same outcomes to the 'moving average'. This indicates a higher unemployment rate for both Portland and the USA during 2009-2011 compared to 2020. How we can address this?

## Yes, we need the smoothing these plots becasue of noise reduction: 

    The decision to apply smoothing techniques like moving averages depends on the objective of the analysis and the characteristics of the data. Smoothing can help in revealing underlying trends and patterns by reducing noise or short-term fluctuations. However, it's essential to consider the trade-offs:

    Noise Reduction: Smoothing techniques can help in reducing noise, making it easier to identify long-term trends in the data. This is evident when examining the unemployment rates for Portland and the USA, which resemble outliers, like to singular instances of noise, notably stemming from the impact of the Covid-19 pandemic during April 2020.

    Loss of Detail: Smoothing might result in a loss of detailed information, especially when using larger window sizes for averaging.

    Impact on Interpretation: Smoothing can potentially alter the perception of the data, emphasizing trends while obscuring short-term variations or sudden changes.

    To assess the need for smoothing and the impact of changes:

    Compare Visualizations: Compare the smoothed and original plots side by side. Does the smoothing reveal clearer trends while preserving essential variations?

    Identify Patterns: Look for patterns or trends that were previously hidden in the noisy data. Are these patterns more evident after applying smoothing?

    Consider Context: Consider the context of the analysis. Are you interested in short-term variations or long-term trends? Does smoothing align with your analysis goals?

    Evaluate Changes: Assess if the application of smoothing has significantly altered the interpretation of the data. Have there been any unexpected shifts or changes in trends due to the smoothing process?

    By comparing the smoothed and original plots and considering the context of the analysis, we can determine whether smoothing is necessary and whether it enhances the understanding of the data.

# Fit time series datasets

    Fitting time series data and making predictions involves several steps, typically including model selection, training, and forecasting. Here's an example using a simple ARIMA (AutoRegressive Integrated Moving Average) model for both Portland and the USA unemployment rate data:

### Firstly, let's fit an ARIMA model to the Portland and USA unemployment rate time series data:

```python
from statsmodels.tsa.arima.model import ARIMA

# Set frequency for both datasets explicitly
portland_unemp_rate.index.freq = 'MS'
usa_unemp_rate.index.freq = 'MS'

# Fit ARIMA model for Portland Unemployment Rate
model_portland = ARIMA(portland_unemp_rate['Portlands Unemployment Rate'], order=(5,1,0))
model_portland_fit = model_portland.fit()

# Fit ARIMA model for USA Unemployment Rate
model_usa = ARIMA(usa_unemp_rate['USA Unemployment Rate'], order=(5,1,0))
model_usa_fit = model_usa.fit()
```

# <span style="color:red;">Prediction Values for August 2025</span>

### Next, let's use these fitted models to predict the unemployment rates for August 2025:

```python
# Predicting unemployment rates for August 2025
august_2025 = pd.to_datetime('2025-08-01')

# Predict for Portland
prediction_portland = model_portland_fit.get_forecast(steps=1).predicted_mean
print(f"Predicted Portland Unemployment Rate for August 2025: {prediction_portland.values[0]}%")

# Predict for USA
prediction_usa = model_usa_fit.get_forecast(steps=1).predicted_mean
print(f"Predicted USA Unemployment Rate for August 2025: {prediction_usa.values[0]}%")
```

## Please note:

    * The orders (p,d,q) in the ARIMA model (e.g., order=(5,1,0)) are examples and may need adjustment based on the data and model performance.
    
    * For more accurate predictions, hyperparameter tuning, model evaluation, and considering more sophisticated models might be necessary.
    
    * The prediction values for August 2025 are based on the fitted ARIMA models and the forecast for one step ahead.
    
    * These predictions provide an estimation for the unemployment rates in August 2025 based on the fitted ARIMA models for Portland and the USA. Adjustments in model parameters or selection of different models may further refine these predictions.
